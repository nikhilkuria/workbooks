```python
from skimage import data
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load and display an image
astronaut = data.astronaut()
plt.imshow(astronaut)
```

```python
# See the shape and size of an image
print('Shape', astronaut.shape)
print('Size', astronaut.size)
```

```python
# Also some stat information
print(astronaut.min(), astronaut.max(), astronaut.mean())
```

```python
# We can also work with the pixel values and modify them
for x in range(200):
    for y in range(len(astronaut[x])):
        astronaut[x][y] = [0,0,0]
plt.imshow(astronaut)
astronaut = data.astronaut()
```

```python
# Or in a smarter way
astronaut[:50] = 0
plt.imshow(astronaut)
astronaut = data.astronaut()
```

```python
# Set a mask to an image, plus great way to play with indexes
# Mask can be any boolean array with the same shape of the image
astronaut_mask = astronaut < 20
astronaut[astronaut_mask] = 255
plt.imshow(astronaut)
astronaut = data.astronaut()
```
